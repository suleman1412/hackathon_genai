```python
import cv2

# Load an image from file
img = cv2.imread('/kaggle/working/cars2_ot1.jpg')

# Print original size
print(f'Original size: {img.shape[:2]}')  # (height, width)

width = 224
height = 224
# Resize the image
new_size = (width, height)  # e.g., (128, 128)
resized_img = cv2.resize(img, new_size)

# Save the resized image
cv2.imwrite('resized.jpg', resized_img)

# Print new size
print(f'Resized size: {resized_img.shape[:2]}')  # (height, width)
```

```python
# NVIDIA NIM INIT

import requests, base64

invoke_url = "https://ai.api.nvidia.com/v1/vlm/microsoft/kosmos-2"

with open("/kaggle/working/cars2_ot1.jpg", "rb") as f:
    image_b64 = base64.b64encode(f.read()).decode()

# print(image_b64)
headers = {
  "Authorization": f"Bearer {API_KEY}",
  "Accept": "application/json"
}

payload = {
  "messages": [
    {
      "role": "user",
      "content": f'Who is in this photo? <img src="data:image/jpg;base64,{image_b64}" />'
    }
  ],
  "max_tokens": 1024,
  "temperature": 0.20,
  "top_p": 0.20
}

response = requests.post(invoke_url, headers=headers, json=payload)
result = response.json()
# response.status_code
print(result['choices'][0]['message']['content'])
```
